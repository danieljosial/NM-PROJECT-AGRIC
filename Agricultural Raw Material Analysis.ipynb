```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("/content/drive/MyDrive/agricultural_raw_material.csv")
```

```python
df.shape
```

```python
df.head()
```

```python
df = df.replace('%', '', regex=True)
df = df.replace(',', '', regex=True)
df = df.replace('-', '', regex=True)
df = df.replace('', np.nan)
```

```python
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
lst = ["Coarse wool Price", "Coarse wool price % Change", "Copra Price", "Copra price % Change", "Cotton price % Change","Fine wool Price", "Fine wool price % Change", "Hard log price % Change", "Hard sawnwood price % Change", "Hide price % change", "Plywood price % Change", "Rubber price % Change", "Softlog price % Change", "Soft sawnwood price % Change", "Wood pulp price % Change"]
df = df.replace('', np.nan)
df = df.replace(' ', '')
df[lst] = df[lst].astype('float')
df.dtypes
```

```python
df.Month  = pd.to_datetime(df.Month.str.upper(), format='%b%y', yearfirst=False)
# Indexing month
df = df.set_index('Month')
df.head()
```

```python
max_price = df.max()
min_price = df.min()
```

```python
print("High-range raw materials: \n",max_price)
print("Low-range raw materials: \n",min_price)
```

```python
df_change = df.pct_change()
max_change = df_change.max()
min_change = df_change.min()
```

```python
print("High %Change materials:\n",max_change)
print("Low %Change materials:\n",min_change)
```

```python
price_range = df.max() - df.min()
print("Range of prices changed over the years:\n", price_range)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
corr = df.corr()
fig = plt.figure(figsize = (100, 12))
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr,mask=mask, annot=True, square=True,cmap='coolwarm')
plt.show()
```
